```python
"""
This script will scrap data from a website with html tags and store it in a csv file using Pandas and BeautifulSoup.
"""
```

```python
import requests
import pandas
from bs4 import BeautifulSoup
```

```python
base_url = "https://www.century21.com/real-estate/flagstaff-az/LCAZFLAGSTAFF/?p="
```

```python
l = []
```

```python
for page in range(1, 4):
    print(base_url + str(page))
    r = requests.get(base_url + str(page))
    c = r.content
    soup = BeautifulSoup(c, "html.parser")
    all = soup.find_all("div", {"class": "property-card"})

    for item in all:
        d = {}
        try:
            d["Address"] = item.find("div", {"class": "property-address"}).text.strip()
        except:
            d["Address"] = None
        try:
            d["City"] = item.find("div", {"class": "property-city"}).text.strip()
        except:
            d["City"] = None
        try:
            d["Beds"] = item.find("div", {"class": "property-beds"}).text.strip()
        except:
            d["Beds"] = None
        try:
            d["Baths"] = item.find("div", {"class": "property-baths"}).text.strip()
        except:
            d["Baths"] = None
        try:
            d["Area"] = item.find("div", {"class": "property-sqft"}).text.strip()
        except:
            d["Area"] = None
        d["Price"] = item.find("a", {"class": "listing-price"}).text.replace("\n", "").strip()
        l.append(d)
```

```python
df = pandas.DataFrame(l)
df.to_csv("Output.csv")
```
